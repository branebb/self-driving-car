```python
from vehicle.vehicle import newCar
#some global settings
FPS = 60
dt = 1.0 / FPS




def animate(car : newCar, time : float):
    #fixed
    frames = int(time / dt)
    car_positions = []
    car_velocities = []
    car_throttle = []
    car_breake = []
    car_gears = []
    car_rpms = []

    for i in range(frames):
        car_positions.append(car.position.copy())
        car_velocities.append(car.velocity)
        car_throttle.append(car.throttle)
        car_breake.append(car.brake)
        car_gears.append(car.gear)
        car_rpms.append(car.rpm)

        if i == frames // 6:
            car.throttle_force = 1.0
            # car.steering_force = 1.0
        
        if i == 2 * frames // 3:
            car.brake_force = 1.0
            car.steering_force = 0.0
        
        car.update_position(dt)


    return np.array(car_positions), np.array(car_velocities), np.array(car_throttle), np.array(car_breake), np.array(car_gears), np.array(car_rpms)   

    
```

```python
car = newCar(position=(1.0, 1.0), heading=0.0, velocity=0.0)
car_pos, car_vels, car_throttle, car_brake, car_gears, car_rpms = animate(car=car, time=20.0)

plt.plot(car_pos[:, 0], car_pos[:, 1])
plt.title("Car Path")
plt.show()
plt.plot(car_vels)
plt.title("Car Velocity")
plt.show()
plt.plot(car_throttle)
plt.title("Car Throttle")
plt.show()
plt.plot(car_brake)
plt.title("Car Brake")
plt.show()
plt.plot(car_gears)
plt.title("Car Gears")
plt.show()
plt.plot(car_rpms)
plt.title("Car RPMs")   
plt.show()
car_vels[-1]
```
